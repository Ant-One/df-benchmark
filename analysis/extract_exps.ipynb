```python
import pandas as pd
import os
import json
import glob
```

```python
dirs = ["ConfDF_norm_v2_all", "ConfDF_norm_v2_th", "ConfDF_norm_v2_th-bb", "ConfDF_norm_v2_th-ob", "ConfDF_norm_v2_th-m"]
all_jsons = []

for dataset in os.scandir("/home/antoine/df-benchmark/exps"):
    if dataset.name in dirs:
        dataset_jsons = list(set(glob.glob(os.path.join(dataset, "**/*.json"), recursive=True)) - set(glob.glob(os.path.join(dataset, "**/preds/**/*.json"), recursive=True)))
        for e in dataset_jsons:
            all_jsons.append(e)
```

```python
all_jsons
```

```python
res_dicts_list = []

for path in all_jsons:
    with open(path) as f:
        res = json.load(f)

        train_dataset = res["train_dataset"]

        if train_dataset == "train_on_ff-orig":
            res["train_dataset"] = "ff-orig"
        elif train_dataset == "train_on_df40-fr-ff":
            res["train_dataset"] = "df40-fr-ff"
        elif train_dataset == "train_on_df40-fs-ff":
            res["train_dataset"] = "df40-fs-ff"
        elif train_dataset == "train_on_df40-all-ff":
            res["train_dataset"] = "df40-all-ff"

        res_dicts_list.append(res)
```

```python
df_results = pd.DataFrame(res_dicts_list)
```

```python
with open("/home/antoine/df-benchmark/analysis/metrics_confdf.csv", "w+") as f:
    df_results.to_csv(f)
```
